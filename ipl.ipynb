```python
import pandas as pd
import numpy as np
import pickle
```

```python
df = pd.read_csv('ipl.csv')
```

```python
df.head()
```

```python
col_to_remove = ['mid','batsman','bowler','striker','non-striker']
df.drop(col_to_remove,axis=1,inplace=True)
```

```python
# only keeping the teams currently playing in ipl
df['bat_team'].unique()

teams = ['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals',
       'Mumbai Indians','Kings XI Punjab','Royal Challengers Bangalore', 
          'Delhi Daredevils', 'Pune Warriors', 'Sunrisers Hyderabad',]

df2 = df[(df['bat_team'].isin(teams)) & df['bowl_team'].isin(teams)]
```

```python
# checking the shapes before and after we have dropped 
df2.shape,df.shape
```

```python
# only keeping data after 5 overs completed
df3 = df2[df2['overs']>=5]
```

```python
df3.shape
```

```python
df3['date'] = pd.to_datetime(df3['date'])
```

```python
df3['date'].dtype
```

```python
df3.reset_index(inplace=True)
```

```python
df3.head()
```

```python
encoded_df = pd.get_dummies(data=df3,
                            columns =['bat_team','bowl_team','venue']
                           )
```

```python
pd.set_option('display.max_columns',None)
encoded_df.head()
```

```python
df_train = encoded_df[encoded_df['date'].dt.year <= 2016]
df_test = encoded_df[encoded_df['date'].dt.year>= 2017]
```

```python
x_train = df_train.drop('total',axis=1)
y_train = df_train.total
x_test = df_test.drop('total',axis=1)
y_test = df_test.total
```

```python
x_train.drop('index',axis=1,inplace=True)
x_test.drop('index',axis=1,inplace=True)
```

```python
x_train.drop('date',axis=1,inplace=True)
x_test.drop('date',axis=1,inplace=True)
```

```python
x_train.info()
```

```python
print(x_train.shape,x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
lin_reg = LinearRegression()
for_reg = RandomForestRegressor(n_estimators=200,max_depth=9,random_state=1,n_jobs=-1)
```

```python
lin_reg.fit(x_train,y_train)
for_reg.fit(x_train,y_train)
```

```python
print(lin_reg.score(x_test,y_test))
print(for_reg.score(x_test,y_test))
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge = Ridge()
parameters = {'alpha':[1e-15,1e-10,1e-8,1e-2,1,5,10,20,30,35,40]}
ridge_regressor = GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
```

```python
ridge_regressor.fit(x_train,y_train)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
prediction = ridge_regressor.predict(x_test)
```

```python
import seaborn as sns
sns.distplot(y_test-prediction)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
lasso = Lasso()
parameters = {'alpha':[1e-15,1e-10,1e-8,1e-2,1,5,10,20,30,35,40]}
lasso_regressor = GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(x_train,y_train)
```

```python
print(ridge_regressor.score(x_test,y_test))
print(lasso_regressor.score(x_test,y_test))
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_test,lin_reg.predict(x_test)))
print(r2_score(y_test,for_reg.predict(x_test)))
print(r2_score(y_test,ridge_regressor.predict(x_test)))
print(r2_score(y_test,lasso_regressor.predict(x_test)))
```

```python
def predict_final_score(batting_team,
                        bowling_team,venue,runs,
                        wickets,overs,runs_last_five,wickets_last_five):

    bat_team_index = np.where(x_train.columns=='bat_team_'+batting_team)[0][0]
    bowl_team_index = np.where(x_train.columns=='bowl_team_'+bowling_team)[0][0]
    venue_index =np.where(x_train.columns=='venue_'+venue)[0][0]
    x = np.zeros(len(x_train.columns))
    x[0] = runs
    x[1] = wickets
    x[2] = overs
    x[3] = runs_last_five
    x[4] = wickets_last_five
    
    if bat_team_index >= 0:
        x[bat_team_index] = 1
    if bowl_team_index >= 0:
        x[bowl_team_index] = 1
    if venue_index > 0:
        x[venue_index] >= 1
    x = x.reshape(1,54)
    return np.round(lasso_regressor.predict(x)[0])
```

```python
predict_final_score('Chennai Super Kings','Kings XI Punjab','Dr DY Patil Sports Academy',120,3,15.2,45,0)
```
